```python
import requests
from bs4 import BeautifulSoup
import json


# Step 1: Send a GET request to the website
url = 'https://viva.bhc.si/strokovnjaki-odgovarjajo/'
question_index = 3000

#make a json array that will store all the questions and answers
output_json = []

for i in range(1, question_index):
    print('Scraping page ' + str(i))

    
    question_url = url + str(i) + '/'
    response = requests.get(question_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    #check if question index exists, check if there is "message error" div
    if soup.find('div', class_='message error'):
        print('--- Page ' + str(i) + ' does not exist.')
        continue
    
    # Find the div element with class "question"
    question_div = soup.find('div', class_='question')
    answer_div = soup.find('div', class_='answer')
    time_div = soup.find('div', class_='time')
    specialization_div = soup.find('div', class_='specialization')
    title_h1 = soup.find('h1', class_='title')
    # <div class="title"><a href="https://viva.bhc.si/strokovnjaki/44/Sanja-Rozman-dr-med">Sanja Rozman, dr. med.</a></div>
    

    # Remove all tags within the div element keep the text content
    for tag in question_div.find_all():
        tag.unwrap()
        
    for tag in answer_div.find_all():
        tag.unwrap()

    # Extract the text content from the div element (now without paragraph tags)
    question_text = question_div.get_text(strip=True)
    answer_text = answer_div.get_text(strip=True)
    
    #make a json object with question, answer, time, specialization and title and append it to the output_json array
    output_json.append({
        'url': question_url,
        'question': question_text,
        'answer': answer_text,
        'time': time_div.get_text(strip=True),
        # 'specialization': specialization_div.get_text(strip=True),
        'title': title_h1.get_text(strip=True)
    })



#save the output_json array to a file use chars   
with open('viva.json', 'w', encoding='utf-8') as f:
    json.dump(output_json, f, ensure_ascii=False, indent=4)
```
