```python
import json
import os
import matplotlib.pyplot as plt

# Adjust font size
plt.rcParams.update({'font.size': 20})

files = ["med-over-net-data-final.json", "viva-data-final.json", "vizita-si-data-final.json"]
forumName = ["MedOverNet", "Viva", "Vizita"]
#list all the colors
# colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'orange', 'cyan', 'magenta', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink', 'lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightgrey', 'lightpink']
colors = ['lightcoral', 'lightskyblue', 'lightgreen']
counts = []

for i in range(len(files)):
    file = open(os.path.join("final-data", files[i]), "r")
    data = json.load(file)
    file.close()
    counts.append(len(data))
    print(forumName[i], len(data))

# Drawing pie chart
plt.figure(figsize=(11, 6))
#align the pie chart to the right
plt.pie(counts, colors=colors, startangle=140, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Creating legend left top
plt.legend(forumName, loc="best", fontsize=16)  # Increase legend font size


plt.show()
```
